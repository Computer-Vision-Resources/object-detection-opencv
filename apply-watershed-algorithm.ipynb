**Import library's**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Create display function**

```python
def display(img, cmap='gray'):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

**Load the image**

```python
sep_coins = cv2.imread("/Users/neemiasbsilva/Downloads/Computer-Vision-with-Python/DATA/pennies.jpg")
```

```python
display(sep_coins)
```

## Segment this images using the simple techniques

**Apply median BLUR**

```python
sep_blur = cv2.medianBlur(sep_coins, 25)
```

```python
display(sep_blur)
```

**Convert to Grayscale**

```python
gray_sep_coins = cv2.cvtColor(sep_blur, cv2.COLOR_BGR2GRAY)
```

```python
display(gray_sep_coins)
```

**Apply Binary Thresholding**

```python
ret, sep_thresholding = cv2.threshold(gray_sep_coins, 160, 255, cv2.THRESH_BINARY_INV)
```

```python
display(sep_thresholding)
```

**Find Contours**

```python
image, contours, hierarchy = cv2.findContours(sep_thresholding.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for i in range(len(contours)):
    
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(sep_coins, contours, i, (255, 0, 0), 10)
```

```python
display(sep_coins)
```

## Segment this images using the Watershed Algorithm

**Load the image**

```python
img = cv2.imread("/Users/neemiasbsilva/Downloads/Computer-Vision-with-Python/DATA/pennies.jpg")
```

**Apply median BLUR**

```python
img = cv2.medianBlur(img,35)
```

```python
display(img)
```

**Convert to Grayscale**

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
display(gray)
```

**Apply Binary Thresholding add OTSU**

```python
ret, threshold = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

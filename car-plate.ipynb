# Using OpenCv to Detect Russian Plate Car

**Import the usual libraries**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Load the Plate Car Image**

```python
img = cv2.imread("/Users/neemiasbsilva/Downloads/Computer-Vision-with-Python/DATA/car_plate.jpg")
# change the BGR to RGB Channels
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

**Creating a function that display the image in a large scale**

```python
def display(img):
    fig = plt.figure(figsize=(12, 12))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
display(img)
```

**Load the haarcascade russian plate number**

```python
russian_cascade = cv2.CascadeClassifier("/Users/neemiasbsilva/Downloads/Computer-Vision-with-Python/DATA/haarcascades/haarcascade_russian_plate_number.xml")
```

**Creating the function to draw the rectangle around of plate**

```python
def detect_plate(img):
    
    plate = img.copy()
    
    plate_rectangles = russian_cascade.detectMultiScale(plate, scaleFactor=1.2, minNeighbors=5)
    
    for (x, y, w, h) in plate_rectangles:
        
        cv2.rectangle(plate, (x, y), (x+w, y+h), (255, 0, 0), 5)
    
    return plate
```

**Show the result**

```python
result = detect_plate(img)
display(result)
```

**Apply the blurs over plate**

```python
def detect_and_blur_plate(img):
    
    plate = img.copy()
    roi = img.copy()
    
    plate_rectangles = russian_cascade.detectMultiScale(plate, scaleFactor=1.2, minNeighbors=5)
    
    for (x, y, w, h) in plate_rectangles:
        
        roi = roi[y:y+h, x:x+w]
        
        blurred_roi = cv2.medianBlur(roi, 7)
        
        plate[y: y+h, x:x+w] = blurred_roi
        
    return plate
```

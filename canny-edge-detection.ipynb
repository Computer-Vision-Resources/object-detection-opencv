# Using Canny Edge Detection Process

###### For more detailed about the concepts of the Canny Detection, please, check: <a href="https://en.wikipedia.org/wiki/Canny_edge_detector">Click Here</a>.

**Import some Libraries**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Load the image**

```python
img = cv2.imread("/Users/neemiasbsilva/Downloads/Computer-Vision-with-Python/DATA/sammy_face.jpg")
```

```python
plt.imshow(img)
# Note: for edge detection you don't need transform BGR to RGB
```

**Apply Canny Edge Detection**

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges) # This edges have a lot of noise
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=127)
```

```python
plt.imshow(edges) #More Noises
```

**Eliminating Noises of Edges**

```python
# calculat the median of values
med_value = np.median(img)
med_value
```

```python
# Lower Threshold to either 0 or 70% of the median value whichever is greater
lower = int(max(0, 0.7 * med_value)) 
# Upper Threshold to either 130% of the median or the max 255, whichever is smaller
upper = int(min(255, 1.3 * med_value))
```

# Harris Corner Detector Applied in Chess

###### To know more about the concept of this corner detector, please check: <a href="https://en.wikipedia.org/wiki/Harris_Corner_Detector">Click Here</a>

**Import library's**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Load images and show**

```python
flat_chess = cv2.imread('/Users/neemiasbsilva/Downloads/Computer-Vision-with-Python/DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_flat_chess, cmap='gray')
```

```python
real_chess = cv2.imread('/Users/neemiasbsilva/Downloads/Computer-Vision-with-Python/DATA/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_real_chess, cmap='gray')
```

```python
gray_flat_chess
```

```python
gray = np.float32(gray_flat_chess)
```

```python
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.03)
```

```python
dst = cv2.dilate(dst, None) 
```

```python
# This process is for visualization
flat_chess[dst > 0.01 * dst.max()] = [255, 0, 0] # RGB
```

```python
plt.imshow(flat_chess)
```

**Make this process in the real chess board image**

```python
gray = np.float32(gray_real_chess)

dst = cv2.cornerHarris(src= gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
dst
```

```python
# This process is for visualization
real_chess[dst > 0.01 * dst.max()] = [255, 0, 0] # RGB
```

```python
plt.imshow(real_chess)
```
